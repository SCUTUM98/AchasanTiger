```python
# Guui-dong map visual
guui_data = pd.read_excel("/Users/bongeungu/Downloads/Guui_xy.xlsx")
location_data = guui_data[['y','x']].values[:len(guui_data)].tolist()

center = [37.544,127.093]
m = folium.Map(location=center, zoom_start=15)

for i in range(len(guui_data)):
    guui_id = guui_data['address_name'][i]
    cnt = int(guui_data['count'][i])
    
    if(cnt >20):
        tmp_cnt = 20
    else:
        tmp_cnt = cnt
    
    if(cnt>30):
        folium.Circle(location = location_data[i], radius = tmp_cnt, color = 'blue', fill = True, fill_opacity = 0.1, tooltip = guui_id, popup = cnt).add_to(m)
    else:
        folium.Circle(location = location_data[i], radius = tmp_cnt, color = 'red', fill = False, fill_opacity = 0.1, tooltip = guui_id, popup = cnt).add_to(m)
m
```
